```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
%matplotlib inline

# Task 1. Load CSV file of COVID-19 cases into a dataframe using pd.read_csv()
cases_df = pd.read_csv('enhanced_sur_covid_19_eng.csv')  # REPLACE WITH READING IN 
cases_df.head()
cases_df.tail()
```

```python
# Task 3.  Show pie charts of status of cases and the classification
st_stat = cases_df.groupby(['Hospitalised/Discharged/Deceased']).count()
st_stat = st_stat[:]['Case no.']
print(st_stat)
fig, ax = plt.subplots(2, figsize=(15, 12))
ax[0].pie(st_stat, labels=st_stat.index, autopct='%1.1f%%', explode=(0.5, 0,
0.1, 1.2, 3.0))
ax[0].set_title('Status of Cases') 

# ADD YOUR CODE to plot to ax[1]
# calculate the frequency of each category in "HK/Non-HK resident" column
res_stat = cases_df.groupby(['HK/Non-HK resident']).count()
res_stat = res_stat[:]['Case no.']
print(res_stat)

# plot the pie chart for "HK/Non-HK resident" column
ax[1].pie(res_stat.tolist(), labels=res_stat.index, autopct='%1.1f%%', explode=(0,0,0))
ax[1].set_title('HK/Non-HK Resident')
```

```python
# Task 4. Create a new column called ‘Asymptomatic’ and set it to ‘Y’ if the case is asymptomatic and ‘N’ if otherwise. 

# ADD YOUR CODE HERE
def setAsymptomatic(row):
    if row['Date of onset'] == 'Asymptomatic':
        return 'Y'
    else:
        return 'N'
    
cases_df['Asymptomatic'] = cases_df.apply(setAsymptomatic, axis=1)
cases_df.tail()
```
